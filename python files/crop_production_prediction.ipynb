```python
features = ['State_Name', 'District_Name', 'Season', 'Crop', 'Area']
X = df[features]
y = df[['Production']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
le4 = LabelEncoder()
X_train['State_Name'] = le1.fit_transform(X_train['State_Name'])
X_train['District_Name'] = le2.fit_transform(X_train['District_Name'])
X_train['Season'] = le3.fit_transform(X_train['Season'])
X_train['Crop'] = le4.fit_transform(X_train['Crop'])
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler_x = StandardScaler()
scaler_y= StandardScaler()
scaler_x.fit(X_train)
X_train = scaler_x.transform(X_train)
X_train = pd.DataFrame(X_train)
```

```python
X_train
```

```python
scaler_y.fit(y_train)
y_train = pd.DataFrame(scaler_y.transform(y_train))
y_train
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(random_state=0)
reg.fit(X_train, y_train)
```

```python
X_test['State_Name'] = le1.transform(X_test['State_Name'])
X_test['District_Name'] = le2.transform(X_test['District_Name'])
X_test['Season'] = le3.transform(X_test['Season'])
X_test['Crop'] = le4.transform(X_test['Crop'])
```

```python
X_test = pd.DataFrame(scaler_x.transform(X_test))
y_pred = reg.predict(X_test)
```

```python
y_test = pd.DataFrame(scaler_y.transform(y_test))
```

```python
y_pred = pd.DataFrame(y_pred)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print("MAE = ", mae)
print("MSE = ", mse)
```

```python
import numpy as np
print("RMSE = ", np.sqrt(mse))
print("R2 = ", r2_score(y_test, y_pred))
```
